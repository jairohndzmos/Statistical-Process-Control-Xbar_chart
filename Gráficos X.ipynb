```python
import pandas as pd
import matplotlib.pyplot as plt
# data = {'Values': [0.08, 0.09, 0.09, 0.1, 0.08, 1.01, 0.09, 0.08, 0.09, 0.09]} 

data = {'Values': [0.25, 0.24, 0.31, 0.28, 0.3, 0.27, 0.33, 0.25, 0.31, 0.32]} 

df = pd.DataFrame(data)
  
df['Obs'] = range(1,len(df['Values'])+1)
```

```python
def x_chart(dataframe):

    A2 = [0,0, 1.880, 1.023, 0.729, 0.577, 0.483, 0.419, 0.373, 0.337, 0.308]
    X_bar_bar = dataframe['Values'].mean()
    R_bar_bar = dataframe['Values'].max() - dataframe['Values'].min()
    A_2 = A2[dataframe['Values'].count()]
    LSC = X_bar_bar + A_2 * R_bar_bar
    LIC = X_bar_bar - A_2 * R_bar_bar
    
    colors = []
    elements_to_rid = []
    for i in dataframe['Values']:
        if i > LSC or i < LIC:
            colors.append('r')
            elements_to_rid.append(i)
        else:
            colors.append('b')
    dataframe.plot(kind='scatter',x='Obs',y='Values',color=colors)
    plt.xticks(range(1, dataframe['Values'].count()+1))
    plt.axhline(y=LSC, color='black', linestyle='-')
    plt.axhline(y=LIC, color='black', linestyle='-')
    plt.axhline(y=X_bar_bar, color='red', linestyle='-')
    plt.plot(dataframe['Obs'], dataframe['Values'], color = 'b')
    
    return plt.show(), elements_to_rid
    
```

```python
def rid_data(dataframe, elements_to_rid):
    
    for i in elements_to_rid:
        dataframe = dataframe.drop(dataframe.index[dataframe['Values'] == i].tolist())
        dataframe['Obs'] = range(1,len(dataframe['Values'])+1)
    
    return dataframe
        
```

```python
#first iteration
p, k = x_chart(df)
print(k)
```

```python
df = rid_data(df, k)
```

```python
#second iteration
p, k = x_chart(df)
print(k)
```

```python
df = rid_data(df, k)
```

```python
#third iteration, on control
p, k = x_chart(df)
print(k)
```
